```
# 1. Import different required library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```
# 2. Check version of the library

print(np.__version__)
print(sns.__version__)
print(pd.__version__)
```

```
# 3. Go for different observation like Info, description

df = pd.read_csv('/content/creditcard.csv')
print(df.info())
df.describe()
```

```
# 4. check null values

df.isnull().sum().sum()
```

```
# 5. Check duplicate values

duplicate = df[df.duplicated()] 
  
print("Duplicate Rows :") 
  
# Print the resultant Dataframe 
duplicate 
```

```
# 6. Filter for different condition - come up with insights from the dataset, basic plot, distribution plots

class_names = {0:'Not Fraud', 1:'Fraud'}
print(df.Class.value_counts().rename(index = class_names))

# There are 239 fraud cases

# Correlation matrix 
corrmat = df.corr() 
fig = plt.figure(figsize = (12, 9)) 
sns.heatmap(corrmat, vmax = .8, square = True) 
plt.show() 

# We can look at an individual feature in Seaborn through a boxplot
sns.boxplot(x="Class", y="V1", data=df)

sns.FacetGrid(df, hue="Class", size=6) \
   .map(sns.kdeplot, "V1") \
   .add_legend()

df.boxplot(by="Class", figsize=(12, 6)) 
```

```
# 7. Use a histogram to plot different columns
df.hist(figsize = (20, 20))
plt.show()
```
